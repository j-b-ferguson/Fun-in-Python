<font size="3" color="#000000"  face="cambria"> <B> A logarithmic spiral is mathematical curiosity first described by Descartes. It is most naturally described in polar coordinates $r$ and $\phi$ by the equation $r(\phi) = e^{k\phi}$ for $\phi \in$ &#8477;.
    

In the natural world, many objects can be modelled as a logarithmic spiral - such as the arms of a spiral galaxy, or the bands of a tropical cyclone.
<br/>
